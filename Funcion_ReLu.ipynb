**Grafica de la funcion ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ReLU(x):
  return np.maximum(0,x)
```

```python
x = np.linspace(-5, 5, 100)
y = ReLU(x)

plt.plot(x, y)
plt.title('Función ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.show()
```
